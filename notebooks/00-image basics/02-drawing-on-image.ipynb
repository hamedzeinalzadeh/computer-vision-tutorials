# Drawing on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(500,500,3),dtype=np.int16)
print ('blank_img shape:', blank_img.shape)
plt.imshow(blank_img)
```

### Shapes
#### Rectangles

- `img`: Image on which the rectangle is to be drawn.
- `pt1`: One vertex of the rectangle(top left).
- `pt2`: The vertex opposite to pt1, defining the rectangle(bottom right).
- `color`: The color or brightness of the rectangle in a grayscale image.
- `thickness`: It forms the rectangle. Negative values (e.g., #FILLED) indicate that the function should draw a filled rectangle.
- `lineType`: The type of the line used for drawing (e.g., solid, dashed). Refer to `#LineTypes` for options.
- `shift`: The number of fractional bits in the point coordinates, used for sub-pixel accuracy.

```python
# This is an inplace funciton
cv2.rectangle(blank_img,pt1=(100,100),pt2=(200,200),color=(255,0,0),thickness=7)
plt.imshow(blank_img)
```

#### Cirlces

```python
cv2.circle(img=blank_img, center=(100,100), radius=50, color=(255,255,0), thickness=5)

#filled in 
cv2.circle(img=blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)

plt.imshow(blank_img)
```

#### Lines

```python
cv2.line(blank_img,pt1=(0,0),pt2=(400,100),color=(102, 255, 255),thickness=4)
plt.imshow(blank_img)
```

#### Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='TEXT',org=(10,400), fontFace=font,fontScale= 4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

#### Polygons

Before drawing a polygon, prepare a vertex array with coordinates. The array should have a shape of `ROWS x 1 x 2`, where ROWS is the number of vertices, and the data type should be **int32**.

```python
# Create a blank image again
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices = np.array([[100,400],[200,400],[200,200],[100,100]],np.int32)
pts = vertices.reshape((-1,1,2))
print ('vertices shape:', vertices.shape, '\npts shape:', pts.shape)
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(0,255,0),thickness=5)

# This is for drawing filled in polylgons
# cv2.fillPoly(blank_img,[pts],color=(0,255,0))

plt.imshow(blank_img)
```
