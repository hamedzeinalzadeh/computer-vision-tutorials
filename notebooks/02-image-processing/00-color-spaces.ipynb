# Color Spaces
Usually we are working with **RGB** color spaces, and in this coding, colors are modeled as a combination of Red, Green and Blue. 

There are another alternative model named **HSL**(hue, saturation, lightness) and **HSV**(hue, saturation, value). HSL and HSV are more closely aligned with the way human vision percieves the color.

Check out this link for more details. https://en.wikipedia.org/wiki/HSL_and_HSV

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00_quail.jpg')
plt.imshow(img)
```

### Converting to RGB

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Converting to HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

### Converting to HSL

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
