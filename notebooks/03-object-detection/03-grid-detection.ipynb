# Grid Detection
Often cameras can create distortion like radial or tangentical, in an image. 
[lens distortion link](https://www.image-engineering.de/library/image-quality/factors/1062-distortion#:~:text=There%20are%20three%20types%20of,be%20removed%20from%20the%20camera.)

Having a **recognizable pattern** attached to the object being tracked, is a good way of accounting distortion while performing operations like object detection. 

**Grid patterns** are often used to calibrate cameras and track motion. the gride should be **specifically** made to **look like chess-board grid**, and then paste to whatever you want to calibrate your camera with.

---
<br>

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../DATA/flat_chessboard.png')
plt.imshow(flat_chess)
```

```python
# the pattern-size is (7, 7) because the most outer grids can not be detected
found, corners = cv2.findChessboardCorners(flat_chess,(7,7))

if found:
    print('OpenCV was able to find the corners')

    flat_chess_copy = flat_chess.copy()
    cv2.drawChessboardCorners(flat_chess_copy, (7, 7), corners, found)
    
else:
    print("OpenCV did not find corners. Double check your patternSize.")
```

```python
plt.imshow(flat_chess_copy)
```

### Circle Based Grid

```python
dots = cv2.imread('images/dot_grid.png')
plt.imshow(dots)
```

```python
found, corners = cv2.findCirclesGrid(dots, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)

if found:
    print('OpenCV was able to find the corners')

    dots_copy = dots.copy()
    cv2.drawChessboardCorners(dots_copy, (10,10), corners, found)
    plt.imshow(dots_copy)
else:
    print("OpenCV did not find corners. Double check your patternSize.")
```
