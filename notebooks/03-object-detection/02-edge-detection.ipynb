### Canny Edge Detector
It is a multi-stage algorithm and one of the most popular ones in edge detection algorithms.

Here is the process in brief:
- Apply Gaussian filter to smooth the image to remove the noise.(You can use your own smoothing and own blurring technique like average kernel)
- Find the intensity gradients of the image.
- Apply non-maximum supression to get rid of spurious edge detection responses.
- Apply double threshold to determine potential edges.
- Track edge by hysterisis: Finalize the edge detecion process by supressing the edges that are weak to connect to strong edges.

> **Note:**<br>
> For **high resolution images**, in purpose of detecting more general edges, it is a good idea to apply your **custom blur** before applying Canny algorithms.


The Canny algorithms also requires a **user** to decide on **low and high threshold values**.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/sun_flowers.jpeg')
plt.imshow(img)
```

### Choosing Threshold

https://stackoverflow.com/questions/25125670/best-value-for-threshold-in-canny

```python
# Calculate the median pixel value
med_val = np.median(img) 

# Lower bound is max between 0 or 70% of the median value
lower = int(max(0, 0.7* med_val))

# Upper bound is the min between 255 or 30% above the median value
upper = int(min(255,1.3 * med_val))

# Blurring the image
blurred_img = cv2.blur(img,ksize=(5,5))

edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)

plt.imshow(edges)
```

```python
# Playing around with thresholds (regarding your specific dataset and your final goal)
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper + 150)

plt.imshow(edges)
```
