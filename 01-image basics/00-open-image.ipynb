# Opening Images with OpenCV

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# There is no error when you are entering the wrong path.(return None instead!)
img = cv2.imread("some/wrong/path.jpg")
print (img)
```

```python
img = cv2.imread('../DATA/sun_flowers.jpeg')
img
```

### Showing Image

```python
img_bgr = cv2.imread('../DATA/sun_flowers.jpeg')
plt.imshow(img_bgr)
```

OpenCV loads images as **BGR** format, but Matplotlib expects **RGB** format for correct display. To achieve this, you can use `cv2.cvtColor()` or directly manipulate the numpy array.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/sun_flowers.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('../DATA/sun_flowers.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

### Resize Image

```python
# width, height, color channels
img_rgb.shape
```

```python
img =cv2.resize(img_rgb,(500,100))
plt.imshow(img)
```

### Resize Image by Ratio

```python
w_ratio = 0.5
h_ratio = 0.5

new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)

plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

- The command `%matplotlib qt` is used in Jupyter notebooks to display plots in an **external window** using the Qt backend. 

### Flipping Images

```python
# For canceling the effect of "%matplotlib qt" used above!
%matplotlib inline
```
